# Reinforcement Learning for PG
In this project, let's try to write a basic model for reinforcement learning. Let's try to use basic data first, and then add custom features.
We will analyze PG shares

```python
import gym
import gym_anytrading

from stable_baselines.common.vec_env import DummyVecEnv
from stable_baselines import A2C

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# 1. Data preparation

```python
# Read data
df = pd.read_csv("pg.us.txt", parse_dates=["Date"])
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df.set_index('Date', inplace=True)
df.head()
```

Since open interest is always empty, let's throw it out of the sample.

```python
df = df.drop("OpenInt", axis=1)
```

```python
df.shape
```

# 2. Preparing

So we have 12075 trading days. During this period, the company has experienced all sorts of ups and downs. Crises and booming times. We will use 75% of days as a training environment. And we will test on the remaining 25%. As a window we will take the diapason of 5 days. Since actually one working week on exchange is 5 days. In contrast to cryptocurrencies, stocks are traded from Monday through Friday.

```python
# Constants
WEEK = 5
TRAIN = int(len(df)*.75)
```

```python
# init environment
env = gym.make('stocks-v0', df=df, frame_bound=(WEEK,TRAIN), window_size=WEEK)
```

```python
env.signal_features
```

```python
# Check the agent's actions (there should be 2: buying or selling)
env.action_space
```

```python
# Running the agent in an environment with a random sequence of actions
state = env.reset()
while True: 
    action = env.action_space.sample()
    n_state, reward, done, info = env.step(action)
    if done: 
        print("info", info)
        break
        
plt.figure(figsize=(15,6))
plt.cla()
env.render_all()
plt.show()
```

Great, the model lost the deposit. Although in fact we could have bought and held and would have earned more than 5000% interest. 
In the next chapter we will train the model and test it on test data.

# 3. Training and Testing
I will intentionally stop training manually. 
We should aim to get a model that:

1) explain_variance - will tend to 1. 

2) value_loss - tends to 0

```python
env_maker = lambda: gym.make('stocks-v0', df=df, frame_bound=(WEEK,TRAIN), window_size=WEEK)
env = DummyVecEnv([env_maker])
```

```python
model = A2C('MlpLstmPolicy', env, verbose=1) 
model.learn(total_timesteps=1000000)
```

I deliberately stopped at explained_variance 0.442. The model can explain 44% of the variance. This result is very rare in the whole number of observations. Apparently without custom features it is useless to continue training on the original data.

```python
env = gym.make('stocks-v0', df=df, frame_bound=(TRAIN-WEEK,len(df)), window_size=WEEK)
obs = env.reset()
while True: 
    obs = obs[np.newaxis, ...]
    action, _states = model.predict(obs)
    obs, rewards, done, info = env.step(action)
    if done:
        print("info", info)
        break
```

```python
plt.figure(figsize=(15,6))
plt.cla()
env.render_all()
plt.show()
```

# Conclusion:

Our model showed a micro+. This is already better than most speculators (as you know, most still lose). This model can be improved by using custom environment features and by improving the architecture of the model itself. This will be done in subsequent studies.
