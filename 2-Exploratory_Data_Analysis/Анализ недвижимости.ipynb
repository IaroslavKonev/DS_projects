## Предобработка данных

### Пропуски в данных

#### Имена населенных пунктов и этажность домов:

У нас есть пропуски в столбцах locality_name (49 пропусков) и floors_total (86 пропусков), всего у нас 23699 строк в данных. 

Название населенного пунка и его удаленность от районного центра могут серьезно влиять на стоимость жилья (для логичесого заполнения данного столбца у нас недостаточно данных). 

Этажность домов мы так же логически заполнить не можем, у нас почти отсутствуют категориальные данные, использование медианы или среднего скорее всего будут не адекватными - разбросы могут быть слишком большие.

И тот и другой столбец имеют в сумме большое небольшое количество пропусков (меньше 1% от всей выборки).
Думаю наилучшим решением будет выброс пропущенных строк.


```python
#выбрасываем пустые имена населенных пунктов и эитажности домов:
df.dropna(subset = ['locality_name', 'floors_total'], inplace=True)

#проверяем ушли ли пропуски
df.isnull().sum()
```

Отлично в данных столбцах больше нет пропусков.

#### Высота потолков

Высота потолков - следущий тип пропущенных данных, данный фактор может влиять на цену, однако есть ГОСТ и СНиП на строительство домов. И обычно строятся дома и квартиры с высотой потолков от 2,5 до 3 метров. Бывают исключения, по типу домов со вторым светом или многоэтажных квартир, но такого жилья строится не так много. 

Заполним эти пропущенные значения медианой, это позволит убрать выбросы по нестандартным квартирам.

```python
df['ceiling_height']=df['ceiling_height'].fillna(df['ceiling_height'].median())

#проверяем ушли ли пропуски
df.isnull().sum()
```

#### Жилые и кухонные площади

Следующие типы пропусков это Жилые площади и Площади кухонь. 

Из обусловленностей указанных мной в первой части работы - люди могут испытывать сложности в вычилсении жилой площади. Люди заполняя объявления могут указывать общую площадь, считая ее более важным фактором, не выделяю жилую, так как вычисление жилой площади может вызвать определенные сложности вычислений и потребовать дополнительных действий для продавца (а люди, как мы знаем, часто - ленивые :) ) совокупность этих факторов влияет на заполняемость. 

Так же в квартирах-студиях вычисление жилой площади и площади кухонь - не простая задача для обывателя, так как площадь кухонных ниш в студиях рассчитывается архитекторами по спец формулам исходя из площадей. 

Для нас важно, что нет пропусков в суммарных площадях. Пропуски в жилих площадях и площадях кухонь я бы оставил.

#### Апартаменты и балконы

Пропуски в столбцах "апартаменты" и "балконы" могут означать несоответствии типа недвижимости и отсутствие балконов соответственно.

Будем заполнять пропуски известными нам значениями:

```python
df['is_apartment']=df['is_apartment'].fillna(False)
df['balcony']=df['balcony'].fillna(0)

#Проверим пропуски в данных
df.isnull().sum()
```

Скорее всего пропуски в этих данных свзаны с тем, что реальные данные не проходят фильтр значимости. Т.е. водоем может и есть, но он на расстоянии более 3 км, такая же ситуация с парками и аэропортами. Они может и есть, но расстояния на столько значимые, что это перестает как-то влиять на цену. Оставим данные пропуски.

#### days_exposition 

Скорее всего пропуски в этих данных обусловлены тем, что сделка не совершена до сих пор. Такое часто бывает с недвижимостью. Некоторые объекты могут продаваться годами. Оставим данные прпоуски.

### Этаж квартиры:

Напишем функцию по определению этажа квартиры по сехеме (первый, последний, другой):
На вход функция должна принимать этаж квартиры и общую этажность дома и через блок условий определять категорию.

```python
def floors(row):
    '''
    Функция на вход принимает ряд таблицы
    Через блок условий сравнивает значение ряда 'floor' с 1 - "первый этаж", со значением ряда из этажности дома - 
    присваивает 'последний этаж'.
    Если ни одного из условий не сработает то на выход получим значение 'другой'
    
    Всего возможно три значения:
    -первый
    -последний
    -другой
    
    ======================================
    
    Аргументы:
                    row(DataFrame): ряд столбца floor
                    
                
    Возвращает:
                    (String): заполняет ячейку столбца одним из классов
                  
                    
    '''    
    
    floor=row['floor']
    if floor==1:
        return 'первый'
    elif floor==row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
print('Строка документации: \n', floors.__doc__)
```

Создадим столбец "floor_cat", будем заполнять его при помощи apply и написаной выше функции:

```python
df['floor_cat']=df.apply(floors, axis=1)

#проверим новый фрейм
df.head()
```
